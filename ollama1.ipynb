```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings =(
    OllamaEmbeddings(model="llama3:latest") ## by default it will use llama2
)
```

```python
r1 =embeddings.embed_documents(
    [" this is my first class in langchina",
     " and i really like leaering gen ai"]
)
```
